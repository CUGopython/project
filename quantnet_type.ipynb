```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base = "https://quantnet.com/tracker/"
type=[]
for i in range(1,341):
    if i==1:
        url=base
    else:
        url=base+'page-'+str(i)
    response = requests.get(url)
    if response.status_code == 200:
        results = BeautifulSoup(response.content,'lxml')
        all_div=results.find_all('li',  {"class": "structItem-type"})
        for _ in all_div:
            type_tag=_.find('span') 
            type.append(type_tag.get_text().strip('\n').strip())
    else:
        print("Failure")
print(type)
```

#####

```python
name=['Type']
test=pd.DataFrame(columns=name,data=type)
print(test)
test.to_csv('D:/quantnettype.csv')
```
