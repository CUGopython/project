```python
a = quant.index   
for i in range(len(a)):
    origpa = quant["GPA"].iloc[i]
    if str(origpa).endswith('/3.8'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-4])/0.95        
    elif str(origpa).endswith('/4'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-2])
    elif str(origpa).endswith('/4.0'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-4])
    elif str(origpa).endswith('/4.3'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-4])/1.075
    elif str(origpa).endswith('/5'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-2])/1.25
    elif str(origpa).endswith('/5.0'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-4])/1.25
    elif str(origpa).endswith('/5.00'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-5])/1.25
    elif str(origpa).endswith('/7'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-2])/1.75
    elif str(origpa).endswith('/10'):
        if len(origpa)>10:
            quant["GPA"].iloc[i]=float(origpa[:3])
        else:
            quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-3])/2.5
    elif str(origpa).endswith('/10.0'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-5])/3
    elif str(origpa).endswith('/12'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-3])/3
    elif str(origpa).endswith('/20'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-3])/5
    elif str(origpa).endswith('/100'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-4])/25
    elif str(origpa).endswith('/ 100'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-6])/25
    elif str(origpa).endswith('/110'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-4])/27.5
    elif str(origpa).endswith('/120'):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-4])/30
    elif ' ' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif '2:01:00' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'Mediocre' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif str(origpa).startswith('201'):
        quant["GPA"].iloc[i]=np.nan  
    elif '?' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif '-' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif str(origpa).startswith('.'):
        quant["GPA"].iloc[i]=np.nan 
    elif 'n/a' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'na' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'Na' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'N.A.' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'NA' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'N/A' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'no' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif '--' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'n.a' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'Hon' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'Cla' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'cla' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'first' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'Upper' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'n.a' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif '(UK)' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif '1st' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif 'No GPA' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif origpa == '':
        quant["GPA"].iloc[i]=np.nan 
    elif '%' in str(origpa):
        quant["GPA"].iloc[i]=float(str(origpa)[:len(str(origpa))-1])/25
    elif 'WES' in str(origpa):
        quant["GPA"].iloc[i]=float(str(origpa)[:4])
    elif 'wes' in str(origpa):
        quant["GPA"].iloc[i]=float(str(origpa)[:4])
    elif 'WUS' in str(origpa):
        quant["GPA"].iloc[i]=float(str(origpa)[:3])
    elif '(core ' in str(origpa):
        quant["GPA"].iloc[i]=float(str(origpa)[:4])
    elif ',' in str(origpa):
        quant["GPA"].iloc[i]=float(str(origpa).replace(',', '.'))
    elif '+' in str(origpa):
        quant["GPA"].iloc[i]=float(origpa[:3])
    elif 'PhD' in str(origpa):
        quant["GPA"].iloc[i]=float(origpa[5:8])
    elif 'HK' in str(origpa):
        new_value=float(origpa[:4])
        quant["GPA"].iloc[i]=new_value * 0.93
    elif '(8.03)' in str(origpa):
        quant["GPA"].iloc[i]=float(origpa[:3])
    elif '(8.0/10)' in str(origpa):
        quant["GPA"].iloc[i]=float(origpa[:4])
    elif 'German' in str(origpa):
        new_value=float(origpa[:4])
        quant["GPA"].iloc[i]=5-new_value
    elif '6/9' in str(origpa):
        quant["GPA"].iloc[i]="2.7"
    elif '[' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif '~' in str(origpa):
        quant["GPA"].iloc[i]=np.nan   
    elif '2:1' in str(origpa):
        quant["GPA"].iloc[i]=np.nan  
    elif '2:2' in str(origpa):
        quant["GPA"].iloc[i]=np.nan 
    elif float(str(origpa))>10:
        quant["GPA"].iloc[i]=float(str(origpa))/25
    elif float(str(origpa))>4 and float(str(origpa))<=10:
        quant["GPA"].iloc[i]=float(str(origpa))/2.5
    else:
        quant["GPA"].iloc[i]=float(origpa)
```
